```python
tickers = ["AAPL", "MSFT", "NVDA", "GOOG"]
tickers = "', '".join(tickers)

# SQL query with DISTINCT to avoid duplicates
Query = f"""
SELECT DISTINCT d.*, t.*
FROM daily_prices d
LEFT JOIN technical_factors t ON d.date = t.date AND t.symbol = d.symbol
WHERE t.symbol IN ('{tickers}')
AND d.symbol IN ('{tickers}')
AND DATE(d.date) >= '2016-01-01'
AND DATE(d.date) <= '2016-12-31'
ORDER BY d.date
"""

# Execute the query
db = sqlite3.connect(local_path)
df = pd.read_sql_query(Query, db)
db.close()

# Remove duplicate columns
df = df.loc[:, ~df.columns.duplicated()]

print(df['date'].dtype)
# Convert 'date' to datetime
#df['date'] = pd.to_datetime(df['date'])
print(df['date'].dtype)

#df['date'] = df['date'].dt.strftime('%Y%m%d').astype(int)


# Set the index to 'date'
#df = df.set_index("date")




# Convert 'symbol' to categorical codes
df['symbol'] = df['symbol'].astype('category').cat.codes

# Calculate percentage change
df['return'] = df['close'].groupby(df['symbol']).pct_change()
df.dropna(inplace=True)
df['log_return'] = df['return'].apply(lambda x: np.log(1 + x))

df['date_seq'] = df.groupby('symbol')['date'].cumcount()
print(df['date_seq'].max() - df['date_seq'].min())

df
```
